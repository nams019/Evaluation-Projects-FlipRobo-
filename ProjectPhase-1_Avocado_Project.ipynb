```python
#Phase 1 - Project 2
# Avocado Project

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, mean_squared_error

url = "https://raw.githubusercontent.com/dsrscientist/Data-Science-ML-Capstone-Projects/master/avocado.csv"
data = pd.read_csv(url)

print(data.head())
print(data.info())


X_classification = data.drop(columns=['type', 'Date', 'Region'])
y_classification = data['type']

X_train_classification, X_test_classification, y_train_classification, y_test_classification = train_test_split(X_classification, y_classification, test_size=0.2, random_state=42)

clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_classification, y_train_classification)

train_preds_classification = clf.predict(X_train_classification)
test_preds_classification = clf.predict(X_test_classification)

train_accuracy_classification = accuracy_score(y_train_classification, train_preds_classification)
test_accuracy_classification = accuracy_score(y_test_classification, test_preds_classification)
print("Classification Task - Train Accuracy:", train_accuracy_classification)
print("Classification Task - Test Accuracy:", test_accuracy_classification)

X_regression = data.drop(columns=['AveragePrice', 'Date', 'Region'])
y_regression = data['AveragePrice']

X_train_regression, X_test_regression, y_train_regression, y_test_regression = train_test_split(X_regression, y_regression, test_size=0.2, random_state=42)

regression_model = LinearRegression()
regression_model.fit(X_train_regression, y_train_regression)

train_preds_regression = regression_model.predict(X_train_regression)
test_preds_regression = regression_model.predict(X_test_regression)

train_rmse_regression = np.sqrt(mean_squared_error(y_train_regression, train_preds_regression))
test_rmse_regression = np.sqrt(mean_squared_error(y_test_regression, test_preds_regression))
print("Regression Task - Train RMSE:", train_rmse_regression)
print("Regression Task - Test RMSE:", test_rmse_regression)

```
